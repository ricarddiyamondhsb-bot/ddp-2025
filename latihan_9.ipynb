```python

# SOAL NO 1
print("SOAL NO 1")

def celcius_ke_fahrenheit(celcius):
    return (celcius * 9/5) + 32

print(celcius_ke_fahrenheit(0))
print(celcius_ke_fahrenheit(100))


# SOAL NO 2
print("\nSOAL NO 2")

def is_genap(bilangan):
    return bilangan % 2 == 0

print(is_genap(4))
print(is_genap(77))


# SOAL NO 3
print("\nSOAL NO 3")

def nilai(n):
    if n >= 75:
        return "lulus"
    else:
        return "gagal"

print(nilai(80))
print(nilai(60))


# SOAL NO 4 (bilangan ganjil)
print("\nSOAL NO 4")

def bilangan(n):
    for i in range(1, n):
        if i % 2 != 0:
            print(i, end=" ")

bilangan(20)
```
